```python
from pandas import json_normalize
import pandas as pd
import requests
```

```python
pd.set_option('max_rows', 99999)
pd.set_option('max_columns', 99999)
```

```python
headers = {
    'Accept': 'application/json, text/plain, */*',
    'Accept-Language': 'en-US,en;q=0.9,pt-BR;q=0.8,pt;q=0.7',
    'Connection': 'keep-alive',
    'Origin': 'https://light-building.messefrankfurt.com',
    'Referer': 'https://light-building.messefrankfurt.com/',
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'same-site',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36',
    'sec-ch-ua': '^\\^Google',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '^\\^Windows^\\^',
}
```

```python
params = (
    ('language', 'en-GB'),
    ('q', ''),
    ('orderBy', 'name'),
    ('pageNumber', '1'),
    ('pageSize', '5'),
    ('showJumpLabels', 'false'),
    ('findEventVariable', 'LIGHTBUILDING'),
)
```

```python
response = requests.get('https://exhibitorsearch.messefrankfurt.com/service/esb/2.1/search/exhibitor', headers=headers, params=params)
```

```python
response.json()
```

```python
light_building = json_normalize(response.json()['result']['hits'])
```

```python
light_building.columns
```

```python
light_building.loc[0]
```

```python
df_light_building = pd.DataFrame(light_building[[
                                                'exhibitor.name',
                                                'exhibitor.address.street',
                                                'exhibitor.address.city',
                                                'exhibitor.address.zip',
                                                'exhibitor.address.tel',
                                                'exhibitor.address.country.label',
                                                'exhibitor.address.email',
                                                'exhibitor.description.text',
                                                'exhibitor.keyWords',
                                                'exhibitor.logo',
                                                'exhibitor.homepage', 
                                                ]]
                                 )
```

```python
df_light_building
```

```python
len(df_light_building)
```

```python
location = {}
```

```python
for i in range(0,len(df_light_building)):
  location[i] = (response.json()['result']['hits'][i]['exhibitor']['exhibition']['exhibitionHall'][0]['name'],(response.json()['result']['hits'][i]['exhibitor']['exhibition']['exhibitionHall'][0]['stand'][0]['name']))
```

```python
location
```

```python
df_location = pd.DataFrame(location).transpose()
```

```python
df_stand = pd.DataFrame(list(location.items()))
```

```python
df_light_building = df_light_building.join(df_location)
```

```python
df_light_building.rename(columns={
                                  'exhibitor.name': 'Exhibitor' ,
                                  'exhibitor.address.street': 'Address' ,
                                  'exhibitor.address.city': 'City',
                                  'exhibitor.address.zip': 'Zip',
                                  'exhibitor.address.tel': 'Tel',
                                  'exhibitor.address.country.label': 'Country',
                                  'exhibitor.address.email': 'email',
                                  'exhibitor.description.text': 'About',
                                  'exhibitor.keyWords': 'Key Words',
                                  'exhibitor.logo': 'Logo',
                                  'exhibitor.homepage': 'WebSite',
                                  0: 'Hall',
                                  1: 'Stand'
                                  }, inplace = True)
```

```python
df_light_building['Logo'] = ('https://exhibitorsearch.messefrankfurt.com/images/medium/')+df_light_building['Logo']
```
